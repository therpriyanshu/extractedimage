```python
import cv2
import numpy as np
```

```python
# Load the first image
image1 = cv2.imread('/content/Apple_01.jpg')
# Load the second image
image2 = cv2.imread('/content/banana_01.jpg')
```

```python
from google.colab.patches import cv2_imshow
```

```python
# Display the first image
cv2_imshow(image1)

# Display the second image
cv2_imshow(image2)
```

```python


#Extract Fruit Regions:
```

```python
# Define a range for red and green colors (adjust these values accordingly)
lower_red = np.array([0, 0, 100])
upper_red = np.array([100, 100, 255])
lower_green = np.array([0, 100, 0])
upper_green = np.array([100, 255, 100])
```

```python
# Create masks for red and green regions
mask1 = cv2.inRange(image1, lower_red, upper_red)
mask2 = cv2.inRange(image2, lower_green, upper_green)
```

```python
# Apply the masks to the original images to extract the fruit regions
fruit1 = cv2.bitwise_and(image1, image1, mask=mask1)
fruit2 = cv2.bitwise_and(image2, image2, mask=mask2)
```

```python
from google.colab.patches import cv2_imshow
```

```python
# Display the extracted fruit regions
cv2_imshow(fruit1)
cv2_imshow(fruit2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Extract RGB Channel Intensity Values:
```

```python
# Extract RGB values for fruit 1
red_channel_1 = fruit1[:, :, 2]  # Red channel
green_channel_1 = fruit1[:, :, 1]  # Green channel
blue_channel_1 = fruit1[:, :, 0]  # Blue channel
```

```python
# Extract RGB values for fruit 2
red_channel_2 = fruit2[:, :, 2]  # Red channel
green_channel_2 = fruit2[:, :, 1]  # Green channel
blue_channel_2 = fruit2[:, :, 0]  # Blue channel
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(red_channel_2)
cv2_imshow(green_channel_2)
cv2_imshow(blue_channel_2)
```

```python
# Create DataFrames to store the data
data1 = {
    'Label': 'fruit1',
    'Red Channel Mean Intensity': red_channel_1,
    'Green Channel Mean Intensity': green_channel_1,
    'Blue Channel Mean Intensity': blue_channel_2,
}
```

```python
data2 = {
    'Label': 'fruit2',
    'Red Channel Mean Intensity': red_channel_2,
    'Green Channel Mean Intensity': green_channel_2,
    'Blue Channel Mean Intensity': blue_channel_2,
}
```

```python
import pandas as pd
df1 = pd.DataFrame([data1])
df2 = pd.DataFrame([data2])
```

```python
# Save the data to Excel files
df1.to_excel('fruit1_intensity_values.xlsx', index=False)
df2.to_excel('fruit2_intensity_values.xlsx', index=False)
```
